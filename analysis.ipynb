```python
import requests

# Suppose df is your DataFrame
# Get the unique currencies in the DataFrame
unique_currencies = df_predicted_1['currency'].unique()

# Make an API call to get the conversion rates
response = requests.get('https://api.exchangerate-api.com/v4/latest/USD')
rates = response.json().get('rates', {})

# Filter the rates to only include the unique currencies in the DataFrame
conversion_rates = {currency: rates.get(currency, 1) for currency in unique_currencies}

# Define a UDF to convert price to USD
def convert_to_usd(price, currency):
    return price * conversion_rates.get(currency, 1)

# Apply the UDF to the DataFrame
df_predicted_1['price_usd'] = df_predicted_1.apply(lambda row: convert_to_usd(row['price'], row['currency']), axis=1)
df_predicted_1[['currency', 'price', 'price_usd']]
```

```python
import hashlib

titles = df[df['predicted_label'] == 1]['title'].dropna()

# Convert titles to lowercase to ensure case-insensitive matching
titles = titles.str.lower()

# Initialize a dictionary to store hash values
hash_dict = {}

# Function to generate hash value for a string
def generate_hash(title):
    return hashlib.sha256(title.encode()).hexdigest()

# Iterate through titles and generate hash values
for idx, title in enumerate(titles):
    hashed_title = generate_hash(title)
    if hashed_title in hash_dict:
        hash_dict[hashed_title].append(idx)
    else:
        hash_dict[hashed_title] = [idx]

# Find potential duplicate ads based on hash collisions
potential_duplicates = [indices for indices in hash_dict.values() if len(indices) > 1]

# Print potential duplicate ads' indices
print("Indices of potential duplicate ads:")
print(potential_duplicates)
```

```python
import numpy as np

# Function to compute cosine similarity between two vectors
def cosine_similarity(vec1, vec2):
    dot_product = np.dot(vec1, vec2)
    norm_vec1 = np.linalg.norm(vec1)
    norm_vec2 = np.linalg.norm(vec2)
    if norm_vec1 == 0 or norm_vec2 == 0:
        return 0
    else:
        return dot_product / (norm_vec1 * norm_vec2)

# Function to convert a text string to a TF-IDF vector
def text_to_tfidf_vector(text, vocab):
    words = text.split()
    vector = np.zeros(len(vocab))
    for word in words:
        if word in vocab:
            vector[vocab[word]] += 1
    return vector / np.linalg.norm(vector)

# Function to create a vocabulary and TF-IDF vectors for all titles
def create_tfidf_vectors(titles):
    vocab = {}
    for title in titles:
        words = title.split()
        for word in words:
            if word not in vocab:
                vocab[word] = len(vocab)
    tfidf_vectors = []
    for title in titles:
        tfidf_vectors.append(text_to_tfidf_vector(title, vocab))
    return vocab, np.array(tfidf_vectors)

# Create TF-IDF vectors for all titles
vocab, tfidf_vectors = create_tfidf_vectors(titles)

# Threshold for considering titles as duplicates (adjust as needed)
threshold = 0.8

# Create a list to store indices of duplicate ads
duplicate_indices = []

# Calculate cosine similarity between each pair of titles
for i in range(len(tfidf_vectors)):
    for j in range(i + 1, len(tfidf_vectors)):
        similarity = cosine_similarity(tfidf_vectors[i], tfidf_vectors[j])
        if similarity > threshold:
            duplicate_indices.append((i, j))

# Print duplicate ads' indices
print("Indices of duplicate ads:")
print(duplicate_indices)
```
